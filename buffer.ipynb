```python
# buffer.py
import numpy as np

class TrajectoryBuffer:
    def __init__(self):
        self.storage = []

    def store(self, transition):
        # transition = dict with keys: state, plan_text, plan_emb, action, reward, value, logprob, next_state, done
        self.storage.append(transition)

    def clear(self):
        self.storage = []

    def __len__(self):
        return len(self.storage)

    def get_all(self):
        return self.storage
```
