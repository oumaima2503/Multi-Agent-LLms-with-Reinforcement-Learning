```python
from agents.analyzer import Analyzer
from agents.recommender import Recommender
from agents.validator import Validator
from envs.decision_env import DecisionEnv
from buffer import TrajectoryBuffer
from train.magrpo_train import MAGRPOTrainer
from utils.logger import SimpleLogger

def main():
    env = DecisionEnv(obs_dim=10, n_actions=4, max_steps=20)
    analyzer = Analyzer()
    buffer = TrajectoryBuffer()
    recommender = Recommender(obs_dim=env.observation_space.shape[0], plan_emb_dim=128, action_dim=env.action_space.n)
    validator = Validator()
    trainer = MAGRPOTrainer(analyzer, recommender, validator, env, buffer)

    # collect & update loop
    for epoch in range(20):
        trainer.collect_episode(max_steps=env.max_steps)
        trainer.update()
        print(f"Epoch {epoch} done")

if __name__ == "__main__":
    main()
```
